```python
import random

# generate a random number between 1 and 13
def generate_card():
    x = random.randint(1,13)
    if (x == 1):
        return "A"
    elif (x == 11):
        return "J"
    elif (x == 12):
        return "Q"
    elif (x == 13):
        return "K"
    else:
        return str(x)
    
def card_value(x):
    if (x == "A"):
        return 11
    elif (x == "J" or x == "Q" or x == "K"):
        return 10
    else:
        return int(x)

def oneGame():
    cards = []
    cards_sum = 0
    dealer_cards = []
    dealer_sum = 0
    lose = False
    tie = False
    insurance = False

    # generate a random card
    c1 = generate_card()
    cards.append(c1)
    cards_sum += card_value(c1)
    d1 = generate_card()
    dealer_cards.append(d1)
    dealer_sum += card_value(d1)
    c2 = generate_card()
    cards.append(c2)
    cards_sum += card_value(c2)

    print("Your cards are: " + cards[0] + " and " + cards[1])
    print("The dealer's card is: " + dealer_cards[0])

    if (dealer_cards[0] == "A"):
        y = input(print("Do you want insurance?")).lower()
        if (y == "yes"):
            insurance = True



    # check if the user wants to hit or stand
    HoS = input("Would you like to hit or stand? ")
    while (HoS == "hit"):
        c = generate_card()
        cards.append(c)
        cards_sum += card_value(c)
        print("Your cards are: " + ", ".join(cards))
        if (cards_sum > 21):
            print("Bust")
            print("You lose")
            lose = True
            break
        HoS = input("Would you like to hit or stand? ")

    # dealer's turn
    if (not lose):
        while (dealer_sum < cards_sum):
            d = generate_card()
            dealer_cards.append(d)
            dealer_sum += card_value(d)
            print("The dealer's cards are: " + ", ".join(dealer_cards))
            if (dealer_sum > 21):
                print("Dealer busts")
                print("You win")
                lose = False
                break
            if (dealer_sum > cards_sum):
                print("Dealer wins")
                print("You lose")
                lose = True
                break
            if (dealer_sum == cards_sum):
                print("Tie")
                tie = True
                lose = True
                break
    return lose, tie, insurance
```

```python
def play():
    bank = 101
    while (bank > 0):
        bet = int(input("How much would you like to bet? "))
        if (bet <= bank):
            lose, tie, insurance = oneGame()
            if lose == False and insurance == False:
                bank += bet
            elif (lose == False and insurance == True):
                bank += 2 * bet
            elif (tie and insurance == False):
                bank += 0
            elif (tie and insurance == True):
                bank -= bet
            elif (lose == True and insurance == True):
                bank += 0
            else:
                bank -= bet
        else:
            print("You don't have enough money to bet that amount")
        print("You have " + str(bank) + " dollars")

        again = input("Would you like to play again? (yes or no) ").lower()
        if (again == "no"):
            break

    print("Game over")
    print("You have " + str(bank) + " dollars")
    
```

```python
play()
```
